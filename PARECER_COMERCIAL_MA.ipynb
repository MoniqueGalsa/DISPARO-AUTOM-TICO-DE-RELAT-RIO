```python
import matplotlib.pyplot as plt
import numpy as np

# Filtrar o DataFrame total_por_grupo para ter apenas as informações "No prazo" e "Fora do prazo"
total_por_prazo = total_por_grupo[total_por_grupo["Prazo"].isin(["NO PRAZO", "FORA DO PRAZO"])]

# Filtrar as cores
colors = {"NO PRAZO": "#15255d", "FORA DO PRAZO": "#8B0000"} 

# Agrupar novamente por "Área responsável pelo cumprimento" e "Prazo" para calcular as somas
somas_por_area = total_por_prazo.groupby(["Área responsável pelo cumprimento", "Prazo"])["Total"].sum().unstack()

# Calcular o tamanho da figura com base no número de xticks
num_xticks = len(somas_por_area.index)
fig_width = max(10, num_xticks * 1.5)  # Ajuste o multiplicador conforme necessário

# Quebrar os nomes das áreas responsáveis em várias linhas se forem muito longos
def quebrar_nome(nome, max_length=15):
    palavras = nome.split()
    linhas = []
    linha_atual = palavras[0]
    
    for palavra in palavras[1:]:
        if len(linha_atual) + len(palavra) + 1 <= max_length:
            linha_atual += ' ' + palavra
        else:
            linhas.append(linha_atual)
            linha_atual = palavra
    linhas.append(linha_atual)
    
    return '\n'.join(linhas)

# Aplicar a função de quebra de linha aos nomes das áreas responsáveis
somas_por_area.index = [quebrar_nome(nome) for nome in somas_por_area.index]

# Ajustar valores baixos
ajuste_minimo = 10.0
somas_por_area_adjusted = somas_por_area.applymap(lambda x: x + ajuste_minimo if x > 0 else 0)

# Criar um gráfico de barras duplas lado a lado com o tamanho da figura ajustado
ax = somas_por_area_adjusted.plot(kind="bar", stacked=False, figsize=(fig_width, 6), width=0.8, color=[colors[col] for col in somas_por_area.columns])

# Personalizar o gráfico
plt.xlabel("Área Responsável pelo Cumprimento")
plt.ylabel("Quantidade de Casos")
plt.xticks(rotation=45, ha="right")
plt.legend(title="Prazo")

# Remover a linha lateral esquerda
ax.spines['left'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

# Remover os valores do eixo Y
ax.set_yticklabels([])

# Adicionar os valores em cima das barras
for container in ax.containers:
    labels = [f"{h - ajuste_minimo:.0f}" if h > ajuste_minimo else "" for h in container.datavalues]
    ax.bar_label(container, labels=labels, label_type="edge", fontsize=14, color="black", padding=8)  # Ajuste o tamanho dos valores aqui

# Ajustar o limite inferior do eixo y
ax.set_ylim(bottom=0)

# Exibir o gráfico
plt.tight_layout()  # Garante que o gráfico seja ajustado corretamente
plt.savefig(r'Monique\parecer_comercial_ma_grafico.png', dpi=300) # Aumenta a resolução (dpi) para melhor qualidade
plt.show()
```

```python

def envia_email_com_imagem_no_corpo(data_atual_formatada, assunto, mensagem,lista_destinatarios ,imagem_base64 ,arquivo_anexo):
    servidor_smtp = 'smtp-mail.outlook.com'
    porta_smtp = 587

    msg = MIMEMultipart()
    msg['From'] = remetente
    msg['To'] = ', '.join(lista_destinatarios)
    msg['Subject'] = assunto

    # Adicionar o corpo do email (em formato HTML)
    mensagem_html = f"""
    <html>
        <body>
            <p>{mensagem}</p>
            <img src="data:image/png;base64,{imagem_base64}" style="max-width: 100%; height: auto;">
        </body>
    </html>
    """

    msg.attach(MIMEText(mensagem_html, 'html'))
    
    # Anexar planilha
    with open(arquivo_anexo, 'rb') as excel_file:
        excel = MIMEApplication(excel_file.read(), name=f'Relatório Parecer Comercial - MA - {data_atual_formatada}.xlsx')
    msg.attach(excel)
    
    try:
        with smtplib.SMTP(host=servidor_smtp, port=porta_smtp) as smtp:
            smtp.starttls()
            smtp.login(remetente, senha)
            smtp.sendmail(remetente, lista_destinatarios, msg.as_string())
        print('Email enviado com sucesso!')
    except Exception as e:
        print('Erro ao enviar o email:', str(e))

import base64
# Ler a imagem como bytes e converter para base64
with open(r'Monique\parecer_comercial_ma_grafico.png', 'rb') as img_file:
    imagem_base64 = base64.b64encode(img_file.read()).decode('utf-8')

caminho_excel_plan_MA = r'Monique\parecer_ma_tratado.xlsx'
# Enviar o email com a imagem incorporada no corpo
envia_email_com_imagem_no_corpo(data_atual_formatada, assuntoMA, mensagemMA, lista_destinatariosMA, imagem_base64,caminho_excel_plan_MA)

```
